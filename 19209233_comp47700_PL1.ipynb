### **Live coding:** Representing a signal digitally
1. Create a sin wave with a period of 0.5 seconds (i.e. frequency = 2 Hz by the relationship $f=\frac{1}{t}$)
2. Sample the sin wave at 100 samples per second so the sampling frequency $f_s= 100$ Hz.
3. Create an array of 201 sinusoidal signal wave samples
4. Set up an array of time samples using `arange`.
5. Compute the amplitude value $x$ of our wave at each time point $x=sin(2\pi f t)$
6. Plot the wave.

**Notes:** `arange` is a NumPy library method see [documentation](http://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html) for details.

```python
f  = 2.0 # Frequency of sine wave = 2 Hz, (i.e. the singal frequency is 2 Hz so the period of 0.5 seconds)
fs = 100 # Sample the signal at 100 samples per second (i.e. the sampling rate is 100 Hz)

t = np.arange(-1,1+1/fs,1/fs) # sample interval, symmetric for convenience later

print('size of t:',t.size)
print('t:',t)

fig,ax = plt.subplots()

x = np.sin(2*np.pi*f*t)
print('x=sin(t):',t)
ax.plot(t,x,'x--')
# units here are seconds on the x-axis as we are plotting the signal over time
ax.set_xlabel('time (s)',fontsize=16)
# no digigal units here -1 -> +1 scale is amplitude and we are using floating point precision
# in the analogue world this could be a voltage (Volts) or current (Amps) or a representation of pressure (Pascals)
ax.set_ylabel('amplitude',fontsize=16)
```

### **Live coding:** Quantization in Python
An analogue sound is a continuous, infinitely divisable signal, represented as $s(t)$. A discrete time signal, has continuous amplitude resolution but it sampled at discrete times, represented as $s[t]$. Conversely a quantized analogue signal $x_Q(t)$ has discrete amplitute. A digital signal, is both time and amplitude discrete and is denoted with square brackets, $s_D[t]$.  A digitial signal has a sampling rate and a bit depth.
1. Generate an analogue (discrete) signal representation (`x`).
2. Generate the quantized signal representation for the analogue signal (`xQ`).
3. Calculate the quantization error `e` (difference between discrete and quantized).
4. Plot the discrete, quantized, and error in the same figure.

```python
N = 1024  # length of signal

# generate signal
x = np.sin(2*np.pi/N * np.arange(N))
# quantize signal
xi = np.round(3 * x)
xQ = 1/3 * xi
e = xQ - x

# plot (quantized) signals
fig, ax1 = plt.subplots(figsize=(10, 4))
ax2 = ax1.twinx()

ax1.plot(x, 'r', label=r'signal $x[k]$')
ax1.plot(xQ, 'b', label=r'quantized signal $x_Q[k]$')
ax1.plot(e, 'g', label=r'quantization error $e[k]$')
ax1.set_xlabel('k')
ax1.set_ylabel(r'$x[k]$, $x_Q[k]$, $e[k]$')
ax1.axis([0, N, -1.2, 1.2])
ax1.legend()

ax2.set_ylim([-3.6, 3.6])
ax2.set_ylabel('quantization index')
ax2.grid()
```

Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio, display
```

Define Parameters

```python
# Signal parameters
frequency = 440  # Frequency in Hz
duration = 1  # Duration in seconds
sampling_rate = 8000  # Sampling rate in Hz
amplitude_range = 0.5  # Dynamic range of amplitude

# Derived parameters
samples = int(duration * sampling_rate)  # Total number of samples
notch_duration_samples = 200  # Number of samples with zero amplitude
```

Generate Signal

```python
# Time vector for the signal
t = np.arange(samples) / sampling_rate

# Generate a cosine signal
cos_signal = amplitude_range * np.cos(2 * np.pi * frequency * t)

# Generate a sine signal
sin_signal = amplitude_range * np.sin(2 * np.pi * frequency * t)

# Apply notch
cos_signal[:notch_duration_samples] = 0
sin_signal[:notch_duration_samples] = 0
```

Plot Signal

```python
plt.figure(figsize=(15, 6))

# Plot cosine signal
plt.subplot(2, 1, 1)
plt.plot(t, cos_signal, label='Cosine Signal')
plt.title('440 Hz Cosine Wave with a Notch')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Plot sine signal
plt.subplot(2, 1, 2)
plt.plot(t, sin_signal, label='Sine Signal')
plt.title('440 Hz Sine Wave with a Notch')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Instantiate Playback Widgets for Cosine Signal

Frequency Content


Both the cosine and sine signals are fundamentally the same in frequency – they are both 440 Hz tones, which is the musical note A4, often used as a tuning standard. The difference between the sine and cosine is just a phase shift; cosine leads the sine wave by 90 degrees.


Dynamic Range



The dynamic range is maintained at [-0.5, +0.5] for both signals as per the assignment's requirement. This ensures that the signal's amplitude does not exceed the specified range, which is important for avoiding clipping when the signal is converted to sound.




Notch Effect

The notch represents a sudden drop to zero amplitude for a duration of 200 samples. This will create a distinct auditory effect, a momentary silence, within the continuous tone. Such a notch can be considered as a form of amplitude modulation and can be used to encode information or create a rhythmic pattern within a tone.



Auditory Perception
Humans are generally more sensitive to changes in sound, including silence within a continuous tone. Therefore, the notch will be perceived as a brief pause or 'blip' in the tone. This can have various applications in signaling or communication, where such interruptions can be used to convey information.



Potential Real-world Applications
Telecommunications: Notches or similar manipulations could be used in tone-based signaling systems, such as those used in older telephone networks.


Music and Sound Design: Creating rhythmic patterns within sustained notes for compositional purposes.
Digital Communications: Encoding information within an audio signal, similar to how dial-up modems used to operate.
Audio Watermarking: The notch could represent a form of audio watermarking where the presence or absence of such notches could carry metadata about the audio track.
Time-Frequency Analysis
A more advanced analysis could involve looking at the signal in the frequency domain using a Fourier transform or a time-frequency representation like a spectrogram. This would show how the notch affects the distribution of energy across different frequencies over time.



Spectral Analysis
A spectral analysis would reveal that the notch introduces additional frequency components due to the sudden change in amplitude. While the dominant frequency will be 440 Hz, the notch will introduce higher frequency harmonics due to its non-sinusoidal shape.




Additional Considerations
The exact auditory perception of the notch can be affected by the listener's hearing abilities, the playback device's characteristics, and the acoustic properties of the environment where the sound is played.

Import Libraries

```python
import numpy as np
from scipy.io import wavfile
from IPython.display import Audio, display
import matplotlib.pyplot as plt
```

Load the WAV File and Preprocess

```python
# Read the WAV file
snd,sr = librosa.load(h_comp)
samples_2_sec = 2 * sr
slice_signal = snd[0:samples_2_sec]
amplified_signal = slice_signal*4
norm_signal = librosa.util.normalize(amplified_signal)
q = 8
q_signal = np.round(q * norm_signal)
q_signal = 1/q * q_signal
ipd.Audio(slice_signal, rate = sr )
ipd.Audio(amplified_signal, rate = sr )
ipd.Audio(q_signal, rate = sr )
```

Plot the Waveforms

```python
plt.figure(figsize=(14, 4))
librosa.display.waveshow(slice_signal, sr=sr, alpha=0.5, color='cyan')  # Change 'cyan' to your preferred color
plt.title('Sliced Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 4))
librosa.display.waveshow(amplified_signal, sr=sr, alpha=0.5)
plt.title('Amplified Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 4))
librosa.display.waveshow(q_signal[:len(slice_signal)], sr=sr, alpha=0.5,color='red')
plt.title('Quantized Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
